```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load the data into a dataframe & Display the first 5 rows
premier_league_data = pd.read_csv(r"C:\Users\ALAMEEN\Desktop\Desktop\Data Analytics\datasets\Data Analytics Python Project\Premier_League_2020_2021.csv")
premier_league_data.head()
```

```python
# Print a summary of the DataFrame
premier_league_data.info()
```

```python
# Print summary statistics of the DataFrame

premier_league_data.describe()
```

```python
# Count the number of missing values (NaN) in each column of premier_league_data

premier_league_data.isnull().sum()
```

```python
# Create 2 new columns to check for the Minutes/ Match and Gaols/ Match

premier_league_data['MinsPerMatch'] = (premier_league_data['Mins'] / premier_league_data['Matches']).astype(int)
premier_league_data['GoalsPerMatch'] = (premier_league_data['Goals'] / premier_league_data['Matches']).astype(float)
premier_league_data.head()
```

```python
# Total Goals scored in the Season

Total_Goals = premier_league_data['Goals'].sum()
print(Total_Goals)
```

```python
#Total Number of Penalty Goals scored

Total_PenaltyGoals = premier_league_data['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

```python
#Total Number of Penalty Attempted

Total_PenaltyAttempts = premier_league_data['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
# Pie chart for penalties missed vs scored

plt.figure(figsize = (13, 6))
pl_not_scored = premier_league_data['Penalty_Attempted'].sum() - Total_PenaltyGoals
data = [pl_not_scored, Total_PenaltyGoals]
labels = ['Penalties Missed', 'Penalty Scored']
color_palette = sns.color_palette("Paired")
plt.pie(data, labels = labels, colors = color_palette, autopct = '%.0f%%')
plt.show()
```

```python
#Total Number of Unique positions

premier_league_data['Position'].unique()
```

```python
# Players from different nations

np.size((premier_league_data['Nationality'].unique()))
```

```python
# Top 10 Nationalities

nationality = premier_league_data.groupby('Nationality').size().sort_values(ascending=False)

ax = nationality.head(10).plot(kind='bar', figsize=(12, 6), color=sns.color_palette('magma'))

for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')

plt.xlabel('Nationality')
plt.ylabel('Number of Players')
plt.title('Top 10 Nationalities with Most Premier League Players')

# Displaying the plot
plt.show()
```

```python
# Players based on age group

Under20 = premier_league_data[premier_league_data['Age'] <= 20]
age20_25 = premier_league_data[(premier_league_data['Age'] > 20) & (premier_league_data['Age'] <= 25)]
age25_30 = premier_league_data[(premier_league_data['Age'] > 25) & (premier_league_data['Age'] <= 30)]
Above30 = premier_league_data[premier_league_data['Age'] > 30]

# Assuming the following DataFrame exist: Under20, age20_25, age25_30 and Above30

x = np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ["<=20", ">20 & <=25", ">25 & <=30", ">30"]
plt.title('Total Players with Age Groups', fontsize=20)
plt.pie(x, labels=mylabels, autopct = "%.1f%%")
plt.show()
```

```python
# Total under 20 players in each club

players_under_20 = premier_league_data[premier_league_data['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette("cubehelix"))
```

```python
# Average age of players in each club

plt.figure(figsize = (12, 6))
sns.boxplot(x = 'Club', y = 'Age', data = premier_league_data)
plt.xticks(rotation = 90)
```

```python
# Top 10 Assists

top_10_assists = premier_league_data[['Name', 'Club', 'Assists', 'Matches']].nlargest(n = 10, columns = 'Assists')
top_10_assists
```

```python
# Creating a DataFrame to group the total goals scored by each club

Goals_by_clubs = pd.DataFrame(premier_league_data.groupby('Club', as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid", color_codes=True)
ax = sns.barplot(x='Club', y='Goals', data=Goals_by_clubs.sort_values(by="Goals"), palette='rocket', hue='Club', legend=False)
ax.set_xlabel("Club", fontsize=30)
ax.set_ylabel("Goals", fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"] = (20, 8)
plt.title('Plot of Club vs Total Goals', fontsize=20)
plt.show()
```

```python
# Most goals by players

top_10_goals = premier_league_data[['Name', 'Club', 'Goals', 'Matches']].nlargest(n = 10, columns = 'Goals')
top_10_goals
```

```python
# Goals per match

top_10_goals_per_match = premier_league_data[[ 'Club', 'Name', 'GoalsPerMatch', 'Matches', 'Goals']].nlargest(n = 10, columns = 'GoalsPerMatch')
top_10_goals_per_match
```

```python
# Pie Chart - Goals with assist and without assist

plt.figure(figsize = (14, 7))
assists = premier_league_data['Assists'].sum()
data = [Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set2')
plt.pie(data, labels = labels, colors = color, autopct ='%.0f%%')
plt.title('Percentage of Goals with Assists', fontsize = 20)
plt.show()
```

```python
# Top 10 players with most yellow cards

epl_yellow = premier_league_data.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20, 6))
plt.title("Players with the most yellow cards")
c = sns.barplot(x=epl_yellow['Name'], y=epl_yellow['Yellow_Cards'], label='Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
plt.xticks(rotation=45)  # Rotate the tick labels directly
plt.show()
```
